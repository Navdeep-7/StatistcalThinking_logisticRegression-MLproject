```python
#Variance
import numpy as np

team1 = [250, 260, 255, 240, 245]
team2 = [350, 180, 50, 390, 200]

var_team1 = np.var(team1) # var is short for variance - inbuilt function of NumPy
var_team2 = np.var(team2)

print(var_team1)
print(var_team2)

```

```python
std_team1 = np.std(team1)
std_team2 = np.std(team2)

print(std_team1)
print(std_team2)
```

```python
import pandas as pd

url2 = "https://raw.githubusercontent.com/rashida048/Datasets/refs/heads/master/StudentsPerformance.csv"
df = pd.read_csv(url2)

df.head()
```

```python
mean_math = df['math score'].mean()
mean_reading = df['reading score'].mean()
mean_writing = df['writing score'].mean()

print("mean score for math is ", mean_math)
print("mean score for reading is ", mean_reading)
print("mean score for writing is ", mean_writing)
```

```python
#If you find out that your daa has a lot of outliers, then using median instead of mean will be a better measure.
median_math = df['math score'].median()
median_reading = df['reading score'].median()
median_writing = df['writing score'].median()

print("median score for math is ", median_math)
print("median score for reading is ", median_reading)
print("median score for writing is ", median_writing)
```

```python
mode_math = df['math score'].mode()[0]
mode_reading = df['reading score'].mode()[0]
mode_writing = df['writing score'].mode()[0]

print("mode for math is ", mode_math)
print("mode for reading is ", mode_reading)
print("mode for writing is ", mode_writing)
```

```python
#What is range?
range_math = df['math score'].max() - df['math score'].min()
range_reading = df['reading score'].max() - df['reading score'].min()
range_writing = df['writing score'].max() - df['writing score'].min()

print("range for math is ", range_math) #You have very weak students and very bright students
print("range for reading is ", range_reading) #
print("range for writing is ", range_writing)
```

```python
#What is variance? #How MUCH the scores vary, as in the magnitude of the difference
variance_math = df['math score'].var()
variance_reading = df['reading score'].var()
variance_writing = df['writing score'].var()

print("variance for math is ", variance_math) # Low variance means that all your students are scoring the same levels
print("variance for reading is ", variance_reading) #
print("variancee for writing is ", variance_writing)
```

```python
#What is Standard Deviation? #How MUCH the scores vary, as in the magnitude of the difference
std_math = df['math score'].std()
std_reading = df['reading score'].std()
std_writing = df['writing score'].std()

print("std for math is ", std_math) # High STD means that your class has mixture of high and low performers
print("std for reading is ", std_reading) #
print("std for writing is ", std_writing)
```
